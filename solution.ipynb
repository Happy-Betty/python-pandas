1.	Write a Pandas program to compare the elements of two Pandas series. 
Sample Series [4,65,436,3,9], [7,0,3,897,9]


```python
# DATA ANALYSIS WITH PANDAS
#SERIES
# Solution
import pandas as pd
A = pd.Series([4,65,436,3,9])
B= pd.Series([7,0,3,897,9])
print("Comparing the elements of the two pandas series")
print("Is A greater than B? ")
print(A > B)
print()
print("Is A less than B? ")
print(A < B)
print()
print("Is A equal to B? ")
print(A == B)
```

Question 2. Write a Pandas program to add, subtract, multiply and divide two Pandas series. 

```python
#Solution

import pandas as pd
x = pd.Series([2,4,6,8,14])
y = pd.Series([1,3,5,7,9])
add = x + y
subtract = x -y
multiply = x * y
divide = x / y
print("The sum is: ")
print(add)
print()
print("The product is: ")
print(multiply)
print()
print("the difference is: ")
print(subtract)
print()
print("The division is: ")
print(divide)
```
